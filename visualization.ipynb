```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
print("setup data viz complete...")
```

```python
# load data
filepath = 'train.csv'

f1_data = pd.read_csv(filepath)
print("successfully getting the data")
```

```python
# lihat keadaan datanya
f1_data.head()
```

```python
# drop idnya
df = f1_data.drop(['id'], axis=1)

# set width and height figure
plt.figure(figsize=(7,4))

# linechart showing
sns.lineplot(data=df)
```

```python
# masukkan dengan benar agar lebih terstruktur
plt.figure(figsize=(7,4))
sns.scatterplot(data=df, x='LapNumber', y='LapTime (s)')
```

```python
# membuat line plotnya
plt.figure(figsize=(7,3))
sns.lineplot(data=df, x='LapNumber', y='LapTime (s)', hue='Compound')
plt.title('LapTime per Lap, dipisah per Compound')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='PitNextLap')
plt.title('Distribusi Target: PitNextLap')
plt.show()

print(df['PitNextLap'].value_counts(normalize=True))
# distribusi target kebanyakan dari tidak mengambil pitnextlap
```

```python
plt.figure(figsize=(7, 4))
sns.countplot(x='Compound', hue='PitNextLap', data=df, palette='Set1')
plt.xlabel('Compound')
plt.ylabel('Count')
plt.title(f'Count of Pit Stops by Compound')
plt.tight_layout()
plt.show()

table = pd.crosstab(df['Compound'], df['PitNextLap'])
print(table)
```
