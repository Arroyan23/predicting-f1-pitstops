```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
print("setup data viz complete...")
```

```python
# load data
filepath = 'train.csv'

f1_data = pd.read_csv(filepath)
print("successfully getting the data")
```

```python
# lihat keadaan datanya
f1_data.head()
```

```python
# drop idnya
df = f1_data.drop(['id'], axis=1)

# set width and height figure
plt.figure(figsize=(11,6))

# linechart showing
sns.lineplot(data=df)
```

```python
# masukkan dengan benar agar lebih terstruktur
plt.figure(figsize=(7,4))
sns.scatterplot(data=df, x='LapNumber', y='LapTime (s)')
```

```python
# membuat line plotnya
plt.figure(figsize=(7,3))
sns.lineplot(data=df, x='LapNumber', y='LapTime (s)')
```
